Introduction:
    
To find the best neighborhood, I thought about crime rate as an indicator to how safe the neighborhood is.

The metric:
I did not find a dataset that shows the crime rate directly, but I did find the dataset which contains the number arrests at each neighborhood in the pittsburgh area.

Best neighborhood:
Based solely on the number of arrests, the best 5 neighrborhood I found are Central Northside, Regent Square, Mt. Oliver Boro, Troy Hill-Herrs Island and Mt. Oliver Neighborhood.

Conclusion:
Although the number of arrests can to some degree indicate the safty of the neighborhood, for example, 4439 arrests at Central Business District certainly tells us that it might not be a safe place to live. But this might not accurately reflect the safty of the area, since if the neighborhood is in the middle of nowhere, then obviously the number of arrests would be smaller significantly. Based solely on the number of arrests, the best neighorhood overall is Mt. Oliver Neighborhood, with only 2 arrests from 2011 to 2014. 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('Arrests.csv')

# Count the incidents per neighborhood
neighborhood_counts = df['INCIDENTNEIGHBORHOOD'].value_counts()

# Plotting
plt.figure(figsize=(60, 20))  # Adjust the size of your chart
neighborhood_counts.plot(kind='bar')
plt.title('Number of Arrests by Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Number of Arrests')
plt.xticks(rotation=90)  # Rotate labels to make them readable
plt.show()
```
