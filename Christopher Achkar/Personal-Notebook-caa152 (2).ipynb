## Intro

My personal submetric for the final project is which neighborhood has the highest school enrollment. <br> I will be using this dataset from 2015: https://data.wprdc.org/dataset/pittsburgh-american-community-survey-2015-school-enrollment (I used the "all races" dataset). <br> This dataset looks at school enrollment of all types, including elementary school, middle school, high school, and even undergraduate and gradaute programs, but I will just be looking at the enrollment totals. <br> Clearly, having high numbers of school enrollment is important to ensure that people are being properly educated. <br> See the code below:
## Code

```python
# import everything needed
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
enrollment = pd.read_csv("school-enrollment.csv") # I shortened the official csv file name
```

### Look at all neighborhoods
Let's make a graph of the overall data that we are working with.

```python
# make a graph of all neighborhoods' school enrollment
enrollment_column = "Estimate; Enrolled in school:"
plt.figure(figsize=(30, 18))
plt.title("School Enrollment in Western PA Neighborhoods", fontsize=30)
plt.xlabel("Neighborhood", fontsize=30)
plt.ylabel("School Enrollment", fontsize=30)
colors = ["indianred"] * len(enrollment["Neighborhood"])
colors[len(colors) - 15] = "dodgerblue" # highlight Squirrel Hill North (eyeballed it)
plt.bar(enrollment["Neighborhood"], enrollment[enrollment_column], color=colors)
plt.xticks(rotation=45, ha = "right", fontsize=14) # rotate x-axis labels, change horizontal alignment, and font size
plt.tight_layout() # add padding
plt.show()
```

### Zoom in
Above, we can see that Squirrel Hill North (the blue bar) has a very high school enrollment. But, the large number of bars is overwhelimg, so I also made a "zoomed-in" version to see it compared to nearby data points.

```python
# filter DataFrame
neighborhoods_to_plot = ["Spring Garden", "Spring Hill-City View", "Squirrel Hill North", "Squirrel Hill South", "St. Clair"]
filtered_enrollment = enrollment[enrollment["Neighborhood"].isin(neighborhoods_to_plot)]

# plotting
plt.figure(figsize=(10, 6))
plt.title("Zoomed-in School Enrollment")
plt.xlabel("Neighborhood")
plt.ylabel("School Enrollment")
colors = ["indianred"] * 5
colors[2] = "dodgerblue" # highlight Squirrel Hill North
plt.bar(filtered_enrollment["Neighborhood"], filtered_enrollment[enrollment_column], color=colors, width=0.5)
plt.show()
```

### Find neighborhoods with best school enrollment
We can see that Squirrel Hill North has very high enrollment of around 6000, but it is good practice to compare it directly with the other high school-enrollment datapoints.

```python
# find top 5 numerical highest school enrollment
top_5 = enrollment.nlargest(5, enrollment_column)
top_5.loc[:, ["Neighborhood", enrollment_column]]
```

Above, we can see the five neighborhoods with highest school enrollment, with Squirrel Hill North in second place. Now, let's visualize this data.


```python
# make a graph from the top 5
plt.figure(figsize=(10, 8))
plt.title("Neighborhoods With Highest School Enrollment")
plt.xlabel("Neighborhood")
plt.ylabel("School Enrollment")
colors = ["indianred"] * len(top_5)
colors[1] = "dodgerblue" # highlight Squirrel Hill North
plt.bar(top_5["Neighborhood"], top_5[enrollment_column], color=colors, width=0.7)
plt.show()
```

## Analysis
Above, we can again see that Squirrel Hill North is the neighborhood with the the second highest school enrollment, with around 5900, which is very high for this dataset.
