## Introduction:
    
To find the best neighborhood, I thought about crime rate as an indicator to how safe the neighborhood is.

The metric:
I did not find a dataset that shows the crime rate directly, but I did find the dataset which contains the number arrests at each neighborhood in the pittsburgh area.

Best neighborhood:
Based solely on the number of arrests, the best 5 neighrborhood I found are Central Northside, Regent Square, Mt. Oliver Boro, Troy Hill-Herrs Island and Mt. Oliver Neighborhood.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('Arrests.csv')

# Count the incidents per neighborhood
neighborhood_counts = df['INCIDENTNEIGHBORHOOD'].value_counts()

# Plotting
plt.figure(figsize=(60, 20))  
neighborhood_counts.plot(kind='bar')
plt.title('Number of Arrests by Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Number of Arrests')
plt.xticks(rotation=90)  # Rotate labels to make them readable
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data 
neighborhoods = ['Central Business District', 'South Side Flats', 'Carrick', 'East Allegheny', 'Homewood South']
incident_counts = [4439, 3496, 2351, 2217, 2090]

# Creating the bar chart
plt.figure(figsize=(10, 8))
plt.bar(neighborhoods, incident_counts, color='skyblue')

# Adding title and labels
plt.title('Incident Counts by Neighborhood(Worst)')
plt.xlabel('Neighborhood')
plt.ylabel('Incident Count')

# Rotating the x-axis labels for better readability
plt.xticks(rotation=45, ha="right")

# Displaying the plot
plt.tight_layout() 
plt.show()

```
