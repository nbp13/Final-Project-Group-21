```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt 
```

```python
##read file
sex_by_occupation = pd.read_csv("sex-by-occupation.csv") 
```

```python
highest_employment = "Estimate; Total:"
top_5_highest_employment = sex_by_occupation.nlargest(10, highest_employment)
top_5_highest_employment.loc[:, ["Neighborhood", highest_employment]]
```

Above is a table outlining exactly how many people are projected to be employed in each Pittsburgh neighborhood in order from the greatest to the least amount of people. Shadyside, Squirrel Hill South, Brookline, Squirrel Hill North, and Bloomfield all rank as the top 5 best neighborhoods in terms of employment. 

```python
plt.figure(figsize = (10,6))

# Title and labels
plt.title("Hightest Employment in Pittsburgh's Neighborhoods")
plt.xlabel("Neighborhood")
plt.ylabel("Employment Estimated Total")

plt.bar(top_5_highest_employment["Neighborhood"], top_5_highest_employment["Estimate; Male:"], color='yellow')
plt.show()
```

```python
plt.figure(figsize = (10,6))

# Title and labels
plt.title("Hightest Employment in Pittsburgh's Neighborhoods")
plt.xlabel("Neighborhood")
plt.ylabel("Employment Estimated Total")

plt.bar(top_5_highest_employment["Neighborhood"], top_5_highest_employment["Estimate; Female:"], color='green')
plt.show()
```

```python
plt.figure(figsize=(8, 8))

# Title
plt.title("Distribution of Employment in Pittsburgh's Neighborhoods")

# Plotting the pie chart
pie = plt.pie(top_5_highest_employment["Estimate; Total:"], labels=top_5_highest_employment["Neighborhood"], autopct='%1.1f%%', startangle=140, colors=['skyblue', 'orange', 'lightgreen', 'pink', 'yellow'])

# Adding a legend outside the pie chart
plt.legend(pie[0], top_5_highest_employment["Neighborhood"], loc="center left", bbox_to_anchor=(1, 0.5))

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

The above bar graph and pie chart provide a visual representation and comparison of Pittsburgh's top 5 neighborhoods in terms of employment. I wanted to include two different graphs for this comparison to see how these neighborhoods compare to each other. The bar graphs with male and female employment totals show Shadyside to have the most employed men and women out of all Pittsburgh neighborhoods. The group's best neighborhood (Squirrel Hill North) ranked 4th best.
